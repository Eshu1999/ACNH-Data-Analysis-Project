```python
nh_months = [
    'NH Jan','NH Feb','NH Mar','NH Apr','NH May','NH Jun',
    'NH Jul','NH Aug','NH Sep','NH Oct','NH Nov','NH Dec'
]

sh_months = [
    'SH Jan','SH Feb','SH Mar','SH Apr','SH May','SH Jun',
    'SH Jul','SH Aug','SH Sep','SH Oct','SH Nov','SH Dec'
]
```

```python
nh_availability = df[nh_months].notna().sum()
sh_availability = df[sh_months].notna().sum()
```

```python
comparison = pd.DataFrame({
    'NH Available Fish': nh_availability.values,
    'SH Available Fish': sh_availability.values
}, index=[
    'Jan','Feb','Mar','Apr','May','Jun',
    'Jul','Aug','Sep','Oct','Nov','Dec'
])
```

```python
comparison
```

```python
plt.scatter(comparison.index,comparison['NH Available Fish'], color= "skyblue",
                   alpha= 0.5, #transprency
                   s= 100, #size of dots
                   label="NH Available Fish"
           )
plt.scatter(comparison.index,comparison['SH Available Fish'], color= "magenta",
                   alpha= 0.5, #transprency
                   s= 100, #size of dots
                   label="SH Available Fish"
           )
plt.title("Fish Availability in different months in NH and SH")
plt.xlabel("Months")
plt.ylabel("Number of type of fishes available")

plt.legend()
plt.show()
# We can observe that when the availability of fishes decreases in one of the hemisphere the other's availability of fishes increases.
```


# Final Insights & Recommendations

### Key Findings
- Certain concepts such as Zen and Modern consistently achieve higher HHA scores.
- Houses with VFX elements tend to score higher on average.
- Moderate DIY usage correlates with better ratings than extremely low or high DIY usage.

### Recommendations
Players aiming to maximize HHA scores should:
- Focus on high-performing design concepts.
- Strategically include VFX elements.
- Balance DIY usage rather than maximizing it.

### Conclusion
This project demonstrates how structured data analysis can reveal meaningful gameplay strategies.
The same analytical workflow can be applied to real-world business and product datasets.
